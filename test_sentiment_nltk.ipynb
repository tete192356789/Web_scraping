```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import nltk

from nltk import sent_tokenize
from nltk import word_tokenize
from nltk import FreqDist
from nltk.corpus import stopwords


url = 'https://www.gamespot.com/articles/elden-ring-millicent-quest-guide/1100-6501520/'
r = requests.get(url)
soup = BeautifulSoup(r.text)
text = soup.get_text()
```

```python
clean_text = text.replace('\n','')
clean_text = clean_text.replace('/','')
clean_text = ''.join([c for c in clean_text if c != "'"])
clean_text
```

```python
import nltk
from nltk import sent_tokenize
from textblob import TextBlob
from textblob.sentiments import NaiveBayesAnalyzer

sentences = sent_tokenize(clean_text)
sentences[6]

tb = TextBlob(sentences[2],analyzer=NaiveBayesAnalyzer()).sentiment
tb
```

```python
words = word_tokenize(clean_text)
len(words)
```

```python
word_no_punc = []
for i in words:
    if i.isalpha():
        word_no_punc.append(i.lower())
len(word_no_punc)
```

```python
word_no_stop = []
stopwords = stopwords.words('english')
for i in word_no_punc:
    if i not in stopwords:
        word_no_stop.append(i)
len(word_no_stop)
```

```python
fdist = FreqDist(word_no_stop)
fdist.most_common(10)
```
